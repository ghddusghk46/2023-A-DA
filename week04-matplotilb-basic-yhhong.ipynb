```python
import matplotlib.pyplot as pit
import numpy as np

N = 9
y = np.zeros(N)

x1 = np.linspace(0, 10, N, endpoint=True)
x2 = np.linspace(0, 10, N, endpoint=False)
pit.plot(x1, y + 0.5, 'o', label='endpoint=True')
pit.plot(x2, y, 'o', label='endpoint=False')

pit.legend()
pit.ylim([-0.5, 1])
pit.show()
```

```python
import matplotlib.pyplot as pit
import numpy as np

N = 8
y = np.zeros(N)

x1 = np.linspace(0, 10, N, endpoint=True)
x2 = np.linspace(0, 10, N, endpoint=False)
pit.plot(x1, y + 0.5, 'o-', label='endpoint=True')
pit.plot(x2, y, 'o:', label='endpoint=False')

pit.legend()
pit.ylim([-0.5, 1])
pit.show()
```
