```python
import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
next(data)

result = []
for row in data:
    for i in range(4, 52):
        row[i] = row[i].replace(',', "")

    row[4:] = map(int, row[4:])
    result.append(row[10])

f.close()

plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(10, 8))
plt.bar(range(len(result)), result)
plt.xlabel('역')
plt.ylabel('승차 인원수')
plt.grid(True)
plt.show()
```

```python
import csv
f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)
next(data)

s_in = [0] * 24
s_out = [0] * 24

for row in data:
    for i in range(4, 52):
        row[i] = row[i].replace(',', "")

    row[4:] = map(int, row[4:])
    for i in range(24):
        s_in[i] = s_in[i] + row[2 * i + 4]
        s_out[i] = s_out[i] + row[2 * i + 5]

    f.close()

    plt.title('지하철 시간대별 승하차 인원 추이')
    plt.plot(s_in, label='승차')
    plt.plot(s_out, label='하차')
    plt.legend()
    plt.grid(True)
    plt.xticks(range(24), range(4, 28))
    plt.show()
```
