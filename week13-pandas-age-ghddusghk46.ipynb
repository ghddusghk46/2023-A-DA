```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('age.csv', encoding='cp949', index_col=0)

df = df.replace(',', regex=True)

df = df.apply(pd.to_numeric)
df = df.div(df['2021년 11월_계_총인구수'], axis='index')
del df['2021년11월_계_총인구수'], df['2021년11월_계_연령구간인구수']

name = input('원하는 지역의 이름을 입력하세요: ')

a = df.index.str.contains(name)

df2 = df[a]

plt.rc('font', family='Malgun Gothic')
df2.T.plot()
plt.show()
```

```python
# A의 인구 비율에서 B의 인구 비율을 뺀다.
x = df.sub(df2.iloc[0], axis='columns')
x.head(3)
```

```python
# # 차이의 제곱 값을 모두 더한다.
y = np.power(x, 2)
z = y.sum(axis='columns')
z
```

```python
z = z[z[:] != 0]
```

```python
i = z.sort_values().index[:5]
i
```

```python
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
df.loc[i].T.plot(figsize=(10, 6))
plt.xlabel('나이')
plt.ylabel('인구 비율')
plt.grid(True)
plt.show()
```
